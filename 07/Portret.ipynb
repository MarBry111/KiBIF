```python
from scapy.all import *
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
pcap = rdpcap("data_apped.pcap")
```

```python
port = []
seq = []
for pac in pcap:
    print(pac[TCP].sport, pac[TCP].seq)
    port.append(pac[TCP].sport)
    seq.append(pac[TCP].seq)
port = np.array(port)
seq = np.array(seq)
suma = port + seq
```

```python
print(np.min(port))
```

```python
xp = port[2:-1]-port[3:]
yp = port[1:-2]-port[2:-1]
zp = port[:-3]-port[1:-2]

xs = seq[2:-1]-seq[3:]
ys = seq[1:-2]-seq[2:-1]
zs = seq[:-3]-seq[1:-2]


xsum = suma[2:-1]-suma[3:]
ysum = suma[1:-2]-suma[2:-1]
zsum= suma[:-3]-suma[1:-2]
```

```python
plt.scatter(xp,yp)
plt.savefig("numPort2D.png")
```
