```python
import random_gen as rg
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
```

```python
#LCG(n, x, a = 22695477, m = 2**32, c = 1)
m = 2**32
xs = rg.LCG(rg.N, 5, m =m)
```

```python
xs_2 = np.array([xs[:-1],xs[1:]])
plt.scatter(xs_2[0,:], xs_2[1,:])
plt.title('Zależność między kolejnymi wyrazami dla m='+str(m), fontsize = 10)
plt.ylabel(r'$x_{i+1}$')
plt.xlabel(r'$x_i$')
plt.savefig('m'+str(m)+'xi.png', bbox_inches = 'tight',pad_inches = 0)
```

```python
xsb = [format(xi, '#010b')[2:] for xi in xs]
```

```python
output = ''
for xi in xsb:
    output = output + xi 
```

```python
abs(output[:20000].count('1') - output[:20000].count('0') )
```

```python
print('1-bit', output[:20000].count('1')+output[:20000].count('0'))
print('2-bit', output[:20000].count('11')+output[:20000].count('00'))
print('3-bit', output[:20000].count('111')+output[:20000].count('000'))
print('4-bit', output[:20000].count('1111')+output[:20000].count('0000'))
print('5-bit', output[:20000].count('11111')+output[:20000].count('00000'))
print('6-bit', output[:20000].count('111111')+output[:20000].count('000000'))
```

```python
plt.xlim(0,max(xs))
plt.ylim(0,max(xs))
for i in range(len(xs)-1):
    plt.plot(xs[i], xs[i+1], 'bo')
    if(i%10 == 0):
        plt.savefig('animation/'+str(i)+'.png', bbox_inches = 'tight',pad_inches = 0)
plt.plot()
```

```python
plt.figure(figsize=(15,5))
plt.title('Kolejne wygenerowane liczby dla m='+str(m))
plt.plot(xs, 'o', ms = 1)
plt.plot(xs, 'r', lw = 1, alpha = 0.1)
plt.xlabel('indeks liczby')
plt.savefig('m'+str(m)+'.png', bbox_inches = 'tight',pad_inches = 0)
```
